```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Trying to create phase shift patterns

width = 1200
height = 950

I0 = 180
Imod = 179
P = 30
phases = [0, np.pi/2, np.pi, 3*np.pi/2]

x = np.arange(width)
y = np.arange(height)
X,Y = np.meshgrid(x,y)

for i, phi in enumerate(phases, 1):
    I = I0 + Imod*np.cos((2*np.pi/P*X + phi))
    plt.imshow(I, cmap = "gray")
    plt.title(f'{phi}')
    plt.savefig(f"{i}.png")
    plt.show()

for i, phi in enumerate(phases, 1):
    I = (np.cos((2*np.pi/P*X + phi)) + 1) * 255
    cv2.imwrite(f"fringe {i}.png", I)



```
