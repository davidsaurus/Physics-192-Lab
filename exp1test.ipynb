```python
import numpy as np

x1=[827.61,1935.82,872.92,1967.35,1043.78,1816.18,1811.46,1043.57,1804.40]
x2=[829.83,1931.46,874.21,1962.8,1042.76,1812.64,1808.89,1043.03,1803.88]
y = [990.26,955.52,1033.01,916.14,1190.98,1111.2,1243.18,1497,1373.17]

npx1 = np.array(x1)
npx2 = np.array(x2) 
ny = np.array(y)

d = npx1 - npx2
print(d)

b = 20 #mm
f = 3*4032/3.58 #px
z = b*f/d
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

grid_x, grid_y = np.mgrid[min(npx1):max(npx1):200j, min(ny):max(ny):200j]
grid_z = griddata((npx1, ny), z, (grid_x, grid_y), method='cubic')

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(grid_x, grid_y, grid_z, cmap='viridis')
ax.set_xlabel("x (px)")
ax.set_ylabel("y (px)")
ax.set_zlabel("Depth z (mm)")
plt.show()
```

```python
import numpy as np

x1=[285,1174,467, 468,734, 1001, 1000, 889,893, 797, 628, 620]
x2=[372,1250,541, 540, 794, 1069, 1066, 940, 948, 857, 684, 673]
y = [40,59,321, 453, 270, 329, 461, 1282,1007, 1002, 1004, 1267]

npx1 = np.array(x1)
npx2 = np.array(x2) 
ny = np.array(y)

d = npx2 - npx1 #px
print(d)

b = 25.4 #mm
f = 3*4032/3.58 #px
z = b*f/d
print(z)
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

grid_x, grid_y = np.mgrid[min(npx1):max(npx1):100j, min(ny):max(ny):100j]
grid_z = griddata((npx1, ny), z, (grid_x, grid_y), method='cubic')

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(grid_x, grid_y, grid_z, cmap='viridis')
ax.set_xlabel("x (px)")
ax.set_ylabel("y (px)")
ax.set_zlabel("Depth z (mm)")
plt.show()
```

```python
plt.figure()
plt.pcolormesh(grid_x, grid_y, np.zeros_like(grid_x), shading='auto', edgecolor='k', linewidth=0.1)
plt.xlabel('grid_x')
plt.ylabel('grid_y')
plt.title('Grid Structure')
plt.axis('equal')
plt.show()
```

```python
print("grid_x shape:", grid_x.shape)
print("grid_y shape:", grid_y.shape)

print("Sample grid_x row:", grid_x[])
print("Sample grid_y column:", grid_y[:, 0])
```

```python
from mpl_toolkits import mplot3d

X, Y = np.meshgrid(grid_x, grid_y)

Z = np.full((100,100), grid_z)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title('Wireframe Plot')

ax.view_init(50, 60)
plt.show()


```
